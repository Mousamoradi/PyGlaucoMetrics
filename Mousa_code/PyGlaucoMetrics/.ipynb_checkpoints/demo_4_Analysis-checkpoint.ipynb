# Progression MD Slope

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')


# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create ID_Eye identifier
df['ID_Eye'] = df['ID'].astype(str) + '_' + df['Eye']

# Sort by ID_Eye and Date
df = df.sort_values(['ID_Eye', 'Date'])

# Calculate time in years from baseline for each ID_Eye
df['years_from_baseline'] = df.groupby('ID_Eye')['Date'].transform(
    lambda x: (x - x.min()).dt.days / 365.25
)

# Count visits per ID_Eye
visit_counts = df.groupby('ID_Eye').size()
valid_id_eyes = visit_counts[visit_counts >= 3].index

# Filter dataset to only include ID_Eye with >= 3 visits
df_filtered = df[df['ID_Eye'].isin(valid_id_eyes)].copy()

print(f"\n{'='*80}")
print("FILTERING DATA")
print("="*80)
print(f"Total ID_Eye combinations in original dataset: {len(visit_counts)}")
print(f"ID_Eye combinations with >= 3 visits: {len(valid_id_eyes)}")
print(f"Total records for analysis: {len(df_filtered)}")
print(f"Unique patients: {df_filtered['ID'].nunique()}")
print(f"Total eyes analyzed: {len(valid_id_eyes)}")

# Function to calculate MD slope for each ID_Eye using OLS regression
def calculate_md_slope(group_data):
    """
    Calculate MD slope using OLS regression
    Returns slope, p-value, standard error, and R-squared
    """
    try:
        # Ensure we have enough data points
        if len(group_data) < 3:
            return pd.Series({
                'md_slope': np.nan,
                'slope_pvalue': np.nan,
                'slope_stderr': np.nan,
                'slope_rsquared': np.nan,
                'baseline_md': np.nan,
                'latest_md': np.nan,
                'followup_years': np.nan,
                'n_visits': len(group_data)
            })
        
        # Remove rows with missing MD values
        clean_data = group_data.dropna(subset=['md']).copy()
        
        if len(clean_data) < 3:
            return pd.Series({
                'md_slope': np.nan,
                'slope_pvalue': np.nan,
                'slope_stderr': np.nan,
                'slope_rsquared': np.nan,
                'baseline_md': np.nan,
                'latest_md': np.nan,
                'followup_years': np.nan,
                'n_visits': len(clean_data)
            })
        
        # Fit OLS regression model
        model = smf.ols("md ~ years_from_baseline", data=clean_data)
        result = model.fit()
        
        # Extract slope (coefficient for years_from_baseline)
        slope = result.params['years_from_baseline']
        pvalue = result.pvalues['years_from_baseline']
        stderr = result.bse['years_from_baseline']
        rsquared = result.rsquared
        
        # Calculate baseline and latest MD
        baseline_md = clean_data.iloc[0]['md']
        latest_md = clean_data.iloc[-1]['md']
        followup_years = clean_data['years_from_baseline'].max()
        
        return pd.Series({
            'md_slope': slope,
            'slope_pvalue': pvalue,
            'slope_stderr': stderr,
            'slope_rsquared': rsquared,
            'baseline_md': baseline_md,
            'latest_md': latest_md,
            'followup_years': followup_years,
            'n_visits': len(clean_data)
        })
        
    except Exception as e:
        return pd.Series({
            'md_slope': np.nan,
            'slope_pvalue': np.nan,
            'slope_stderr': np.nan,
            'slope_rsquared': np.nan,
            'baseline_md': np.nan,
            'latest_md': np.nan,
            'followup_years': np.nan,
            'n_visits': len(group_data)
        })

# Calculate MD slope for each ID_Eye
print(f"\n{'='*80}")
print("CALCULATING MD SLOPES")
print("="*80)
print("Using OLS regression: md ~ years_from_baseline")
print("This may take a few moments...\n")

slope_results = df_filtered.groupby('ID_Eye').apply(calculate_md_slope).reset_index()

# Split ID_Eye back into ID and Eye
slope_results[['ID', 'Eye']] = slope_results['ID_Eye'].str.split('_', expand=True)
slope_results['ID'] = slope_results['ID'].astype(int)

# Create MD_Slope_progression label based on slope value
def classify_progression(slope):
    if pd.isna(slope):
        return None
    elif slope < -0.5:
        return 'worsening'
    elif slope >= 0:
        return 'improving'
    else:  # -0.5 <= slope < 0
        return 'stable'

slope_results['MD_Slope_progression'] = slope_results['md_slope'].apply(classify_progression)

# Merge slopes back to the original filtered dataframe
df_with_slopes = df_filtered.merge(
    slope_results[['ID_Eye', 'md_slope', 'slope_pvalue', 'slope_stderr', 
                   'slope_rsquared', 'baseline_md', 'latest_md', 
                   'followup_years', 'n_visits', 'MD_Slope_progression']], 
    on='ID_Eye', 
    how='left'
)

# Get the most recent visit for each ID_Eye for summary
latest_visits = df_with_slopes.sort_values('Date').groupby('ID_Eye').last().reset_index()

# Create final results summary
final_results = latest_visits[[
    'ID', 'Eye', 'Age', 'male', 'race', 'yearsfollowed', 
    'md', 'vfi', 'md_slope', 'slope_pvalue', 'slope_stderr', 'slope_rsquared',
    'baseline_md', 'latest_md', 'followup_years', 'n_visits', 'MD_Slope_progression'
]].copy()

# Sort by MD slope (fastest progressors first)
final_results = final_results.sort_values('md_slope')

# Summary statistics
print("="*80)
print("SUMMARY STATISTICS")
print("="*80)
print(f"\nUnique Patients: {final_results['ID'].nunique()}")
print(f"Total Eyes Analyzed: {len(final_results)}")
print(f"  - OD (Right Eyes): {(final_results['Eye'] == 'OD').sum()}")
print(f"  - OS (Left Eyes): {(final_results['Eye'] == 'OS').sum()}")
print(f"\nMD Slope Statistics (dB/year):")
print(f"  Mean: {final_results['md_slope'].mean():.4f}")
print(f"  Median: {final_results['md_slope'].median():.4f}")
print(f"  Std Dev: {final_results['md_slope'].std():.4f}")
print(f"  Min: {final_results['md_slope'].min():.4f}")
print(f"  Max: {final_results['md_slope'].max():.4f}")
print(f"\nModel Fit (R-squared):")
print(f"  Mean: {final_results['slope_rsquared'].mean():.4f}")
print(f"  Median: {final_results['slope_rsquared'].median():.4f}")
print(f"\nFollow-up Duration (years):")
print(f"  Mean: {final_results['followup_years'].mean():.2f}")
print(f"  Median: {final_results['followup_years'].median():.2f}")
print(f"  Range: {final_results['followup_years'].min():.2f} - {final_results['followup_years'].max():.2f}")
print(f"\nProgression Categories (MD slope):")
print(f"  Fast progressors (< -1.0 dB/year): {(final_results['md_slope'] < -1.0).sum()}")
print(f"  Moderate progressors (-1.0 to -0.5 dB/year): {((final_results['md_slope'] >= -1.0) & (final_results['md_slope'] < -0.5)).sum()}")
print(f"  Slow progressors (-0.5 to 0 dB/year): {((final_results['md_slope'] >= -0.5) & (final_results['md_slope'] < 0)).sum()}")
print(f"  Stable/Improving (>= 0 dB/year): {(final_results['md_slope'] >= 0).sum()}")
print(f"\nStatistically Significant Slopes (p < 0.05): {(final_results['slope_pvalue'] < 0.05).sum()}")

# Calculate progressors (slope < -0.5 dB/year)
progressors = final_results[final_results['md_slope'] < -0.5].copy()
print(f"\n{'='*80}")
print(f"PROGRESSORS (MD slope < -0.5 dB/year)")
print(f"{'='*80}")
print(f"Total Eyes with slope < -0.5: {len(progressors)}")
print(f"Unique Patients with slope < -0.5: {progressors['ID'].nunique()}")
print(f"  - Percentage of total eyes: {len(progressors)/len(final_results)*100:.1f}%")
print(f"  - Percentage of total patients: {progressors['ID'].nunique()/final_results['ID'].nunique()*100:.1f}%")

# Count unique ID_Eye per progression label
print(f"\n{'='*80}")
print("UNIQUE ID_EYE COUNT PER MD_SLOPE_PROGRESSION")
print("="*80)
print(f"Worsening (slope < -0.5): {(final_results['MD_Slope_progression'] == 'worsening').sum()}")
print(f"Stable (-0.5 <= slope < 0): {(final_results['MD_Slope_progression'] == 'stable').sum()}")
print(f"Improving (slope >= 0): {(final_results['MD_Slope_progression'] == 'improving').sum()}")

# Display first 20 results
print("\n" + "="*80)
print("FIRST 20 RESULTS (Fastest Progressors)")
print("="*80)
print(final_results[['ID', 'Eye', 'md_slope', 'MD_Slope_progression', 'slope_pvalue', 'slope_rsquared', 
                     'n_visits', 'baseline_md', 'latest_md', 
                     'followup_years', 'Age']].head(20).to_string(index=False))

# Save results to CSV
output_filename = r'Ensemble_with_MD_Slopes_OLS.csv'
df_with_slopes.to_csv(output_filename, index=False)
print(f"\n{'='*80}")
print(f"Complete dataset with slopes saved to: {output_filename}")

# Save summary
summary_output = r'MD_Slopes_Summary_OLS.csv'
final_results.to_csv(summary_output, index=False)
print(f"Summary results saved to: {summary_output}")
print("="*80)
```

```python
from PyVisualFields import vfprogression
import numpy as np

# Randomly select an ID
first_id = np.random.choice(df_VFs['ID'].unique())

# Filter for selected ID
df_VFs1 = df_VFs[df_VFs['ID'] == first_id].copy()

# Convert Date to datetime for proper sorting
df_VFs1['Date'] = pd.to_datetime(df_VFs1['Date'])

# Sort by Date from earliest to latest
df_VFs1 = df_VFs1.sort_values('Date')

# Rename 'Eye' column to 'eyeid' and map 'OD' to 1, 'OS' to 2
df_VFs1 = df_VFs1.rename(columns={'Eye': 'eyeid'})
df_VFs1['eyeid'] = df_VFs1['eyeid'].map({'OD': 1, 'OS': 2})

# Convert eyeid to float64
df_VFs1['eyeid'] = df_VFs1['eyeid'].astype('float64')

# Convert all tdp columns to float64
for i in range(1, 55):
    tdp_col = f'tdp{i}'
    if tdp_col in df_VFs1.columns:
        df_VFs1[tdp_col] = df_VFs1[tdp_col].astype('float64')

# Define the columns to keep (52 TDP values: tdp1-tdp54 excluding tdp26 and tdp35)
tdp_columns = [f'tdp{i}' for i in range(1, 55) if i not in [26, 35]]
columns_to_keep = ['ID', 'yearsfollowed', 'eyeid', 'Date'] + tdp_columns

# Filter the DataFrame to keep only the specified columns that exist
columns_to_keep = [col for col in columns_to_keep if col in df_VFs1.columns]
df_VFs1 = df_VFs1[columns_to_keep]

# Filter to keep only eyeid 1 and 2, then sort by eyeid and Date
df_VFs1 = df_VFs1[df_VFs1['eyeid'].isin([1.0, 2.0])]
df_VFs1 = df_VFs1.sort_values(['eyeid', 'Date']).reset_index(drop=True)

# Check which eyes are present
unique_eyes = df_VFs1['eyeid'].unique()

print(f"Randomly selected ID: {first_id}")
print(f"Shape: {df_VFs1.shape}")
print(f"Columns: {list(df_VFs1.columns)}")
print(f"Number of TDP columns: {len([c for c in df_VFs1.columns if c.startswith('tdp')])}")
print(f"Eyeid distribution:\n{df_VFs1['eyeid'].value_counts().sort_index()}")
print(f"Eyes present: {unique_eyes}")

# Prepare data for progression analysis (without Date column)
df_VFs1_analysis = df_VFs1.drop(columns=['Date'])

# Run CIGTS progression analysis
results = vfprogression.progression_cigts(df_VFs1_analysis)

# Handle results based on which eyes are present
progression_map = {}

if len(results) == 1:
    # Only one eye present
    if 1.0 in unique_eyes:
        progression_map[1.0] = results[0]  # OD only
    elif 2.0 in unique_eyes:
        progression_map[2.0] = results[0]  # OS only
elif len(results) >= 2:
    # Both eyes present: (left eye, right eye) = (OS, OD) = (eyeid 2, eyeid 1)
    if 1.0 in unique_eyes:
        progression_map[1.0] = results[1]  # Right eye (OD) - second result
    if 2.0 in unique_eyes:
        progression_map[2.0] = results[0]  # Left eye (OS) - first result

# Add progression_cigts column
df_VFs1['progression_cigts'] = df_VFs1['eyeid'].map(progression_map)

print(f"\nProgression results: {results}")
print(f"\nData with progression_cigts column (sorted by Date):")
print(df_VFs1[['ID', 'Date', 'yearsfollowed', 'eyeid', 'progression_cigts']].head(20))
```

```python
import pandas as pd
from PyVisualFields import vfprogression
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("PROCESSING ALL IDs FOR CIGTS PROGRESSION")
print("="*80)

# Function to process one ID
def process_id_cigts(df_VFs, patient_id):
    """
    Process CIGTS progression analysis for a single patient ID
    Returns a dictionary with ID, Eye, and progression status
    """
    try:
        # Filter for this ID
        df_VFs1 = df_VFs[df_VFs['ID'] == patient_id].copy()
        
        if len(df_VFs1) == 0:
            return None
        
        # Convert Date to datetime for proper sorting
        df_VFs1['Date'] = pd.to_datetime(df_VFs1['Date'])
        
        # Sort by Date from earliest to latest
        df_VFs1 = df_VFs1.sort_values('Date')
        
        # Rename 'Eye' column to 'eyeid' and map 'OD' to 1, 'OS' to 2
        df_VFs1 = df_VFs1.rename(columns={'Eye': 'eyeid'})
        df_VFs1['eyeid'] = df_VFs1['eyeid'].map({'OD': 1, 'OS': 2})
        
        # Convert to appropriate dtypes
        df_VFs1['eyeid'] = df_VFs1['eyeid'].astype('float64')
        df_VFs1['yearsfollowed'] = df_VFs1['yearsfollowed'].astype('float64')
        
        # Convert all tdp columns to float64
        for i in range(1, 55):
            tdp_col = f'tdp{i}'
            if tdp_col in df_VFs1.columns:
                df_VFs1[tdp_col] = df_VFs1[tdp_col].astype('float64')
        
        # Define the columns to keep (52 TDP values: tdp1-tdp54 excluding tdp26 and tdp35)
        tdp_columns = [f'tdp{i}' for i in range(1, 55) if i not in [26, 35]]
        columns_to_keep = ['ID', 'yearsfollowed', 'eyeid'] + tdp_columns
        
        # Filter the DataFrame to keep only the specified columns that exist
        columns_to_keep = [col for col in columns_to_keep if col in df_VFs1.columns]
        df_VFs1 = df_VFs1[columns_to_keep]
        
        # Filter to keep only eyeid 1 and 2, then sort by eyeid
        df_VFs1 = df_VFs1[df_VFs1['eyeid'].isin([1.0, 2.0])]
        df_VFs1 = df_VFs1.sort_values('eyeid').reset_index(drop=True)
        
        if len(df_VFs1) == 0:
            return None
        
        # Check which eyes are present
        unique_eyes = df_VFs1['eyeid'].unique()
        
        # Run CIGTS progression analysis - suppress R warnings
        import rpy2.rinterface_lib.callbacks as rcb
        
        def no_console(x):
            pass
        
        old_write_console = rcb.consolewrite_print
        rcb.consolewrite_print = no_console
        
        try:
            results = vfprogression.progression_cigts(df_VFs1)
            rcb.consolewrite_print = old_write_console
        except Exception as e:
            rcb.consolewrite_print = old_write_console
            raise e
        
        # Handle results based on which eyes are present
        progression_map = {}
        
        if len(results) == 1:
            # Only one eye present
            if 1.0 in unique_eyes:
                progression_map[1.0] = results[0]  # OD only
            elif 2.0 in unique_eyes:
                progression_map[2.0] = results[0]  # OS only
        elif len(results) >= 2:
            # Both eyes present: (left eye, right eye) = (OS, OD) = (eyeid 2, eyeid 1)
            if 1.0 in unique_eyes:
                progression_map[1.0] = results[1]  # Right eye (OD) - second result
            if 2.0 in unique_eyes:
                progression_map[2.0] = results[0]  # Left eye (OS) - first result
        
        return {'ID': patient_id, 'progression_map': progression_map}
        
    except Exception as e:
        # Silently skip patients with errors
        return None

# Get unique IDs
unique_ids = df_VFs['ID'].unique()
print(f"\nProcessing {len(unique_ids)} unique patients...")

# Process all IDs
all_results = []
errors_count = 0

for idx, patient_id in enumerate(unique_ids):
    if idx % 500 == 0:
        print(f"Processing patient {idx+1}/{len(unique_ids)}... (Successful: {len(all_results)}, Errors: {errors_count})")
    
    result = process_id_cigts(df_VFs, patient_id)
    if result is not None:
        all_results.append(result)
    else:
        errors_count += 1

print(f"\nProcessing complete!")
print(f"Successfully processed: {len(all_results)} patients")
print(f"Skipped due to errors: {errors_count} patients")

# Create a mapping dictionary for easy lookup
progression_lookup = {}
for result in all_results:
    patient_id = result['ID']
    progression_map = result['progression_map']
    for eyeid, progression in progression_map.items():
        # Map eyeid back to OD/OS
        eye = 'OD' if eyeid == 1.0 else 'OS'
        progression_lookup[(patient_id, eye)] = progression

# Add progression_cigts column to original dataframe
df_VFs['progression_cigts'] = df_VFs.apply(
    lambda row: progression_lookup.get((row['ID'], row['Eye']), None), 
    axis=1
)

# Display results
print("\n" + "="*80)
print("RESULTS")
print("="*80)
print(f"Total records in dataset: {len(df_VFs)}")
print(f"Records with progression_cigts assigned: {df_VFs['progression_cigts'].notna().sum()}")
print(f"Records with progression_cigts missing: {df_VFs['progression_cigts'].isna().sum()}")
print(f"\nProgression_cigts distribution:")
print(df_VFs['progression_cigts'].value_counts(dropna=False))

# Count unique ID_Eye per progression label
print("\n" + "="*80)
print("UNIQUE ID_EYE COUNT PER PROGRESSION LABEL")
print("="*80)
df_VFs['ID_Eye'] = df_VFs['ID'].astype(str) + '_' + df_VFs['Eye']
unique_id_eye_labels = df_VFs[['ID_Eye', 'progression_cigts']].drop_duplicates()

print(f"\nTotal unique ID_Eye: {len(unique_id_eye_labels)}")
print(f"\nUnique ID_Eye per progression label:")
print(unique_id_eye_labels['progression_cigts'].value_counts(dropna=False))

# Save the updated dataframe
output_path = r'Ensemble_filtered_with_progression_cigts.csv'
df_VFs.to_csv(output_path, index=False)
print(f"\n{'='*80}")
print(f"Updated dataframe with progression_cigts saved to:")
print(f"{output_path}")
print("="*80)
```

```python
import pandas as pd
from PyVisualFields import vfprogression
import warnings
warnings.filterwarnings('ignore')

# Step 1: Filter for ID_Eye with >= 5 repeats
print("="*80)
print("FILTERING FOR ID_EYE WITH >= 5 REPEATS")
print("="*80)

df_VFs['ID_Eye'] = df_VFs['ID'].astype(str) + '_' + df_VFs['Eye']

# Count repeats per ID_Eye
id_eye_counts = df_VFs.groupby('ID_Eye').size().reset_index(name='n_repeats')
id_eyes_with_3plus = id_eye_counts[id_eye_counts['n_repeats'] >= 5]['ID_Eye'].tolist()

print(f"Total ID_Eye combinations: {len(id_eye_counts)}")
print(f"ID_Eye with >= 5 repeats: {len(id_eyes_with_3plus)}")

# Filter the dataset
df_VFs_filtered = df_VFs[df_VFs['ID_Eye'].isin(id_eyes_with_3plus)].copy()

print(f"Total records before filtering: {len(df_VFs)}")
print(f"Total records after filtering: {len(df_VFs_filtered)}")
print(f"Unique patients after filtering: {df_VFs_filtered['ID'].nunique()}")

# Function to process one ID
def process_id_progression(df_VFs, patient_id):
    """
    Process progression analysis for a single patient ID
    Returns a dictionary with ID, eyeid, and progression status
    """
    try:
        # Filter for this ID
        df_VFs1 = df_VFs[df_VFs['ID'] == patient_id].copy()
        
        if len(df_VFs1) == 0:
            return None
        
        # Rename 'Eye' column to 'eyeid' and map 'OD' to 1, 'OS' to 2
        df_VFs1 = df_VFs1.rename(columns={'Eye': 'eyeid'})
        df_VFs1['eyeid'] = df_VFs1['eyeid'].map({'OD': 1, 'OS': 2})
        
        # Convert eyeid to float64
        df_VFs1['eyeid'] = df_VFs1['eyeid'].astype('float64')
        
        # Convert all td columns to int32 (round to nearest integer first)
        for i in range(1, 55):
            td_col = f'td{i}'
            if td_col in df_VFs1.columns:
                df_VFs1[td_col] = df_VFs1[td_col].round().astype('int32')
        
        # Define the columns to keep (52 TD values: td1-td54 excluding td26 and td35)
        td_columns = [f'td{i}' for i in range(1, 55) if i not in [26, 35]]
        columns_to_keep = ['ID', 'yearsfollowed', 'eyeid'] + td_columns
        
        # Filter the DataFrame to keep only the specified columns that exist
        columns_to_keep = [col for col in columns_to_keep if col in df_VFs1.columns]
        df_VFs1 = df_VFs1[columns_to_keep]
        
        # Filter to keep only eyeid 1 and 2, then sort by eyeid
        df_VFs1 = df_VFs1[df_VFs1['eyeid'].isin([1.0, 2.0])]
        df_VFs1 = df_VFs1.sort_values('eyeid').reset_index(drop=True)
        
        if len(df_VFs1) == 0:
            return None
        
        # Run progression analysis - suppress R warnings
        import rpy2.rinterface_lib.callbacks as rcb
        import rpy2.rinterface as rinterface
        
        # Temporarily suppress R console output
        def no_console(x):
            pass
        
        old_write_console = rcb.consolewrite_print
        rcb.consolewrite_print = no_console
        
        try:
            results = vfprogression.progression_plrnouri2012(df_VFs1)
            rcb.consolewrite_print = old_write_console
        except Exception as e:
            rcb.consolewrite_print = old_write_console
            raise e
        
        # Results are (left eye, right eye) = (OS, OD) = (eyeid 2, eyeid 1)
        # Create mapping for each eye
        progression_map = {}
        if len(results) >= 2:
            progression_map[1.0] = results[1]  # Right eye (OD) - second result
            progression_map[2.0] = results[0]  # Left eye (OS) - first result
        
        return {'ID': patient_id, 'progression_map': progression_map}
        
    except Exception as e:
        # Silently skip patients with errors
        return None

# Get unique IDs from filtered data
unique_ids = df_VFs_filtered['ID'].unique()
print(f"\n{'='*80}")
print(f"PROCESSING {len(unique_ids)} UNIQUE PATIENTS")
print("="*80)

# Process all IDs
all_results = []
errors_count = 0
for idx, patient_id in enumerate(unique_ids):
    if idx % 500 == 0:
        print(f"Processing patient {idx+1}/{len(unique_ids)}... (Successful: {len(all_results)}, Errors: {errors_count})")
    
    result = process_id_progression(df_VFs_filtered, patient_id)
    if result is not None:
        all_results.append(result)
    else:
        errors_count += 1

print(f"\nProcessing complete!")
print(f"Successfully processed: {len(all_results)} patients")
print(f"Skipped due to errors: {errors_count} patients")

# Create a mapping dictionary for easy lookup
progression_lookup = {}
for result in all_results:
    patient_id = result['ID']
    progression_map = result['progression_map']
    for eyeid, progression in progression_map.items():
        # Map eyeid back to OD/OS
        eye = 'OD' if eyeid == 1.0 else 'OS'
        progression_lookup[(patient_id, eye)] = progression

# Add progression_plr column to filtered dataframe
df_VFs_filtered['progression_plr'] = df_VFs_filtered.apply(
    lambda row: progression_lookup.get((row['ID'], row['Eye']), None), 
    axis=1
)

# Merge n_repeats for context BEFORE displaying
df_VFs_filtered = df_VFs_filtered.merge(id_eye_counts[['ID_Eye', 'n_repeats']], on='ID_Eye', how='left')

# Display results
print("\n" + "="*80)
print("RESULTS")
print("="*80)
print(f"Total records in filtered dataset: {len(df_VFs_filtered)}")
print(f"Records with progression_plr assigned: {df_VFs_filtered['progression_plr'].notna().sum()}")
print(f"Records with progression_plr missing: {df_VFs_filtered['progression_plr'].isna().sum()}")
print(f"\nProgression_plr distribution:")
print(df_VFs_filtered['progression_plr'].value_counts(dropna=False))

# Show sample
print(f"\nSample of data with progression_plr:")
print(df_VFs_filtered[['ID', 'Eye', 'yearsfollowed', 'n_repeats', 'progression_plr']].head(20))

# Save the updated dataframe
output_path = 'progression_plr_5vfs.csv'
df_VFs_filtered.to_csv(output_path, index=False)
print(f"\nFiltered dataframe with progression saved to: {output_path}")
```

```python
import pandas as pd
from PyVisualFields import vfprogression
import warnings
warnings.filterwarnings('ignore')

# Function to process one ID
def process_id_schell2014(df_VFs, patient_id):
    """
    Process Schell 2014 progression analysis for a single patient ID
    Returns a dictionary with ID, Eye, and progression status
    """
    try:
        # Filter for this ID
        df_VFs1 = df_VFs[df_VFs['ID'] == patient_id].copy()
        
        if len(df_VFs1) == 0:
            return None
        
        # Convert Date to datetime for proper sorting
        df_VFs1['Date'] = pd.to_datetime(df_VFs1['Date'])
        
        # Sort by Date from earliest to latest
        df_VFs1 = df_VFs1.sort_values('Date')
        
        # Rename 'Eye' column to 'eyeid' and map 'OD' to 1, 'OS' to 2
        df_VFs1 = df_VFs1.rename(columns={'Eye': 'eyeid'})
        df_VFs1['eyeid'] = df_VFs1['eyeid'].map({'OD': 1, 'OS': 2})
        
        # Convert to appropriate dtypes
        df_VFs1['eyeid'] = df_VFs1['eyeid'].astype('float64')
        df_VFs1['md'] = df_VFs1['md'].astype('float64')
        df_VFs1['yearsfollowed'] = df_VFs1['yearsfollowed'].astype('float64')
        
        # Filter to keep only eyeid 1 and 2
        df_VFs1 = df_VFs1[df_VFs1['eyeid'].isin([1.0, 2.0])]
        df_VFs1 = df_VFs1.sort_values(['eyeid', 'Date']).reset_index(drop=True)
        
        if len(df_VFs1) == 0:
            return None
        
        # Check which eyes are present
        unique_eyes = df_VFs1['eyeid'].unique()
        
        # Prepare data for progression analysis
        df_VFs1_analysis = df_VFs1[['ID', 'yearsfollowed', 'eyeid', 'md']].copy()
        
        # Run progression analysis - suppress R warnings
        import rpy2.rinterface_lib.callbacks as rcb
        
        def no_console(x):
            pass
        
        old_write_console = rcb.consolewrite_print
        rcb.consolewrite_print = no_console
        
        try:
            results = vfprogression.progression_schell2014(df_VFs1_analysis)
            rcb.consolewrite_print = old_write_console
        except Exception as e:
            rcb.consolewrite_print = old_write_console
            raise e
        
        # Handle results based on which eyes are present
        progression_map = {}
        
        if len(results) == 1:
            # Only one eye present
            if 1.0 in unique_eyes:
                progression_map[1.0] = results[0]  # OD only
            elif 2.0 in unique_eyes:
                progression_map[2.0] = results[0]  # OS only
        elif len(results) >= 2:
            # Both eyes present: (left eye, right eye) = (OS, OD) = (eyeid 2, eyeid 1)
            if 1.0 in unique_eyes:
                progression_map[1.0] = results[1]  # Right eye (OD) - second result
            if 2.0 in unique_eyes:
                progression_map[2.0] = results[0]  # Left eye (OS) - first result
        
        return {'ID': patient_id, 'progression_map': progression_map}
        
    except Exception as e:
        # Silently skip patients with errors
        return None

# Get unique IDs
unique_ids = df_VFs['ID'].unique()
print(f"\nProcessing {len(unique_ids)} unique patients...")

# Process all IDs
all_results = []
errors_count = 0

for idx, patient_id in enumerate(unique_ids):
    if idx % 500 == 0:
        print(f"Processing patient {idx+1}/{len(unique_ids)}... (Successful: {len(all_results)}, Errors: {errors_count})")
    
    result = process_id_schell2014(df_VFs, patient_id)
    if result is not None:
        all_results.append(result)
    else:
        errors_count += 1

print(f"\nProcessing complete!")
print(f"Successfully processed: {len(all_results)} patients")
print(f"Skipped due to errors: {errors_count} patients")

# Create a mapping dictionary for easy lookup
progression_lookup = {}
for result in all_results:
    patient_id = result['ID']
    progression_map = result['progression_map']
    for eyeid, progression in progression_map.items():
        # Map eyeid back to OD/OS
        eye = 'OD' if eyeid == 1.0 else 'OS'
        progression_lookup[(patient_id, eye)] = progression

# Add progression_schell column to original dataframe
df_VFs['progression_schell'] = df_VFs.apply(
    lambda row: progression_lookup.get((row['ID'], row['Eye']), None), 
    axis=1
)

# Display results
print("\n" + "="*80)
print("RESULTS")
print("="*80)
print(f"Total records in dataset: {len(df_VFs)}")
print(f"Records with progression_schell assigned: {df_VFs['progression_schell'].notna().sum()}")
print(f"Records with progression_schell missing: {df_VFs['progression_schell'].isna().sum()}")
print(f"\nProgression_schell distribution:")
print(df_VFs['progression_schell'].value_counts(dropna=False))

# Show sample
print(f"\nSample of data with progression_schell:")
print(df_VFs[['ID', 'Eye', 'Date', 'md', 'progression_schell']].head(20))

# Save the updated dataframe
output_path = r'Ensemble_filtered_with_progression_schell.csv'
df_VFs.to_csv(output_path, index=False)
print(f"\n{'='*80}")
print(f"Updated dataframe with progression_schell saved to:")
print(f"{output_path}")
print("="*80)
```

```python
import pandas as pd
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr

# Activate automatic conversion between pandas and R dataframes
pandas2ri.activate()

# Function to create eyeid and prepare dataframe
def prepare_vf_dataframe(df):
    """
    Prepare df_VFs dataframe by creating eyeid and ensuring requirements for progression.schell2014.
    
    Parameters:
        df (pd.DataFrame): Input dataframe with columns ID, Eye, md, etc.
        
    Returns:
        pd.DataFrame: Prepared dataframe with eyeid and md columns, filtered for >= 4 measurements per eyeid.
    """
    # Check if required columns exist
    required_columns = ['ID', 'Eye', 'md']
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Missing required columns: {missing_columns}")
    
    # Print input dataframe info
    print(f"Input dataframe shape: {df.shape}")
    print(f"Input dataframe columns: {df.columns.tolist()}")
    
    # Create eyeid: Map OD to 1, OS to 2, and combine with ID
    df['eyeid'] = df.apply(lambda row: f"{row['ID']}_{1 if row['Eye'] == 'OD' else 2}", axis=1)
    
    # Check for missing md values and remove them
    if df['md'].isna().any():
        print(f"Warning: {df['md'].isna().sum()} rows with missing md values removed.")
        df = df.dropna(subset=['md'])
    
    # Print measurement counts before filtering
    measurement_counts = df.groupby('eyeid').size()
    print("\nMeasurements per eyeid before filtering:")
    print(measurement_counts)
    
    # Ensure at least 4 measurements per eyeid
    valid_eyeids = measurement_counts[measurement_counts >= 4].index
    if len(valid_eyeids) < len(measurement_counts):
        print(f"Warning: {len(measurement_counts) - len(valid_eyeids)} eyeids have < 4 measurements and are excluded.")
    if len(valid_eyeids) == 0:
        raise ValueError("No eyeids have >= 4 measurements after filtering.")
    
    df = df[df['eyeid'].isin(valid_eyeids)]
    
    # Select only required columns for progression.schell2014
    measmatrix = df[['eyeid', 'md']]
    
    # Print measurement counts after filtering
    if not measmatrix.empty:
        print("\nMeasurements per eyeid after filtering:")
        print(measmatrix.groupby('eyeid').size())
    else:
        print("Warning: measmatrix is empty after filtering.")
    
    return measmatrix

# Function to run progression.schell2014 in R
def run_progression_schell2014(measmatrix):
    """
    Run progression.schell2014 from vfprogression package in R using rpy2.
    
    Parameters:
        measmatrix (pd.DataFrame): Dataframe with eyeid and md columns.
        
    Returns:
        pd.DataFrame: Results with eyeid and progression status.
    """
    try:
        # Load vfprogression R package
        vfprogression = importr('vfprogression')
        
        # Convert pandas dataframe to R dataframe
        r_measmatrix = pandas2ri.py2rpy(measmatrix)
        
        # Run progression.schell2014 for each unique eyeid
        unique_eyeids = measmatrix['eyeid'].unique()
        results = []
        
        for eyeid in unique_eyeids:
            # Subset dataframe for this eyeid
            r_subset = r_measmatrix[r_measmatrix.rx2('eyeid') == eyeid]
            
            # Debug: Print number of rows for this eyeid
            print(f"Processing eyeid {eyeid} with {len(r_subset.rx2('md'))} measurements")
            
            # Call progression.schell2014
            progression_result = vfprogression.progression_schell2014(r_subset)
            
            # Convert result to Python string
            progression_status = str(progression_result[0])
            
            results.append({'eyeid': eyeid, 'progression': progression_status})
        
        # Create results dataframe
        results_df = pd.DataFrame(results)
        return results_df
    
    except Exception as e:
        print(f"Error running progression.schell2014: {str(e)}")
        return None

# Example usage
if __name__ == "__main__":
    
    try:
        # Prepare dataframe
        measmatrix = prepare_vf_dataframe(df_VFs)
        
        if not measmatrix.empty:
            # Run progression analysis
            results = run_progression_schell2014(measmatrix)
            
            if results is not None:
                print("\nProgression Results:")
                print(results)
        else:
            print("No valid data to process after filtering.")
            
    except ValueError as e:
        print(f"Error: {str(e)}")
```

### glr: Linear regression  (from visualFields package)

```python
'''
performs global linear regression analysis

# type: 
    ‘ms‘: mean sensitivity
    ‘ss‘: standard deviation of sensitivities
    ‘md‘: mean deviation of total deviation values
    ‘sd‘: standard deviation of total deviation values
    ‘pmd‘: pattern mean deviation
    ‘psd‘: pattern standard deviation
    ‘vfi‘: VFI    
    ‘gh‘: general height 
    
# testSlope:
    slope, or slopes, to test as null hypothesis. Default is 0.
    
    
return values:
    
    - id patient ID
    – eye patient eye
    – type type of data analysis. . 
    – testSlope slope for glr or list of slopes for plr to test as null hypotheses
    – nvisits number of visits
    – years years from baseline. Used for the pointwise linear regression analysis
    – data data analyzed. 
    – pred predicted values. Each column is a location of the visual field used for the analysis.
    Each row is a visit (as many as years)
    – sl slopes estimated at each location for pointwise (simple) linear regression
    – int intercept estimated at each location for pointwise (simple) linear regression
    – tval t-values obtained for the left-tailed-t-tests for the slopes obtained in the pointwise
    (simple) linear regression at each location
    – pval p-values obtained for the left-tailed t-tests for the slopes obtained
'''

from PyVisualFields import visualFields
import numpy as np
import matplotlib.pyplot as plt

df_VFs_py = visualFields.data_vfpwgRetest24d2()
df_VFs_py = df_VFs_py.loc[(df_VFs_py.id==1) & (df_VFs_py.eye=='OD')]
df_gi = visualFields.getgl(df_VFs_py) # get global indices
res = visualFields.glr(df_gi, type = "md", testSlope = 0) #linear regression with global indices
print(res.keys())


intercept =  float(res['int'])
slope = float(res['sl'])
se =  float(res['se'])
tval = float(res['tval'])
pval = float(res['pval'])
years = res['years']


x = np.linspace(0, 1, num=50)
y = df_gi['tmd'].values
# Create a list of values in the best fit line
abline_values = [slope * i + intercept for i in x]
plt.plot(x, abline_values, '--')
plt.scatter(years, y)
plt.xlabel('years')
plt.ylabel('TMD')
plt.title('mean deviation of total deviation values')
```

### plr: performs pointwise linear regression (PLR) analysis (from visualFields package)

```python
''' performs pointwise linear regression (PLR) analysis

# type: 
    ‘s‘: sensitivities
    ‘td‘: total deviation values
    ‘pd‘: pattern deviation values
    
# testSlope:
    slope, or slopes, to test as null hypothesis. Default is 0.
    if a single value, then the same null hypothesis is used for all locations.
    If a vector of values, then (for plr
    and poplr) each location of the visual field will have a different null hypothesis.
    The length of testSlope must be 1 or equal to the number of locations to be used
    in the PLR or PoPLR analysis
    
    
return values:
    
    - id patient ID
    – eye patient eye
    – type type of data analysis. . 
    – testSlope slope for glr or list of slopes for plr to test as null hypotheses
    – nvisits number of visits
    – years years from baseline. Used for the pointwise linear regression analysis
    – data data analyzed. 
    – pred predicted values. Each column is a location of the visual field used for the analysis.
    Each row is a visit (as many as years)
    – sl slopes estimated at each location for pointwise (simple) linear regression
    – int intercept estimated at each location for pointwise (simple) linear regression
    – tval t-values obtained for the left-tailed-t-tests for the slopes obtained in the pointwise
    (simple) linear regression at each location
    – pval p-values obtained for the left-tailed t-tests for the slopes obtained
'''

from PyVisualFields import visualFields
import numpy as np


df_VFs_py = visualFields.data_vfpwgRetest24d2()
df_VFs_py = df_VFs_py.loc[(df_VFs_py.id==1) & (df_VFs_py.eye=='OD')]
res = visualFields.plr(df_VFs_py, type='s', testSlope=0) # pointwise linear regression (PLR) 

print('===> keys is res:  \n', res.keys())
intercept = res['int']
slope = res['sl']
standarderror = res['se']
tval = res['tval']
pval = res['pval']
print('===> keys in slope: \n', slope.keys())
arrObejct = np.asarray(list(slope.items()), dtype=object)
slopes_numpy = np.asarray(arrObejct[:,1], dtype=float) 
print('===>slope values:  \n', slopes_numpy)
arrObejct = np.asarray(list(intercept.items()), dtype=object)
intercepts_numpy = np.asarray(arrObejct[:,1], dtype=float) 
print('===> intercepts values:  \n', intercepts_numpy)
```

### poplr: performs PoPLR regression analysis as in O’Leary et al (from visualFields package). 

#### Reference: N. O’Leary, B. C. Chauhan, and P. H. Artes. Visual field progression in glaucoma: estimating the overall significance of deterioration with permutation analyses of pointwise linear regression (PoPLR). Investigative Ophthalmology and Visual Science, 53, 2012

```python
'''
performs PoPLR analysis as in O’Leary et al:
    N. O’Leary, B. C. Chauhan, and P. H. Artes. Visual field progression in glaucoma: estimating
    the overall significance of deterioration with permutation analyses of pointwise linear regression
    (PoPLR). Investigative Ophthalmology and Visual Science, 53, 2012

# type: 
    ‘s‘: sensitivities
    ‘td‘: total deviation values
    ‘pd‘: pattern deviation values
    
# testSlope:
    slope, or slopes, to test as null hypothesis. Default is 0.
    if a single value, then the same null hypothesis is used for all locations.
    If a vector of values, then (for plr
    and poplr) each location of the visual field will have a different null hypothesis.
    The length of testSlope must be 1 or equal to the number of locations to be used
    in the PLR or PoPLR analysis
    
# nperm:
    number of permutations. If the number of visits is 7 or less, then nperm =factorial(nrow(vf)).
# trunc:        
    truncation value for the Truncated Product Method (see reference)
    
    
return values:
    
    - id patient ID
    – eye patient eye
    – type type of data analysis. . 
    – testSlope slope for glr or list of slopes for plr to test as null hypotheses
    – nvisits number of visits
    – years years from baseline. Used for the pointwise linear regression analysis
    – data data analyzed. 
    – pred predicted values. Each column is a location of the visual field used for the analysis.
    Each row is a visit (as many as years)
    – sl slopes estimated at each location for pointwise (simple) linear regression
    – int intercept estimated at each location for pointwise (simple) linear regression
    – tval t-values obtained for the left-tailed-t-tests for the slopes obtained in the pointwise
    (simple) linear regression at each location
    – pval p-values obtained for the left-tailed t-tests for the slopes obtained
    
    – csl the modifed Fisher’s S-statistic for the left-tailed permutation test
    – cslp the p-value for the left-tailed permutation test
    – csr the modifed Fisher’s S-statistic for the right-tailed permutation test
    – csrp the p-value for the right-tailed permutation test
    – pstats a list with the poinwise slopes (‘sl‘), intercepts (‘int‘), standard errors (‘se‘),
    and p-values (‘pval‘) obtained for the series at each location analyzed and for all nperm
    permutations (in ‘permutations‘)
    – cstats a list with all combined stats:
    * csl,csr the combined Fisher S-statistics for the left- and right-tailed permutation
    tests respectively
    * cslp,csrp the corresponding p-values for the permutation tests
    * cslall,csrall the combined Fisher S-statistics for all permutations
    
'''

from PyVisualFields import visualFields
import numpy as np

# Permutation of PLR with TD values
df_VFs_py = visualFields.data_vfpwgRetest24d2()
df_VFs_py = df_VFs_py.loc[(df_VFs_py.id==1) & (df_VFs_py.eye=='OD')]
res = visualFields.poplr(df_VFs_py, type = "td", testSlope = 0, nperm = 'default', trunc = 1) # Permutation of PLR with TD values
print(res.keys())
```
